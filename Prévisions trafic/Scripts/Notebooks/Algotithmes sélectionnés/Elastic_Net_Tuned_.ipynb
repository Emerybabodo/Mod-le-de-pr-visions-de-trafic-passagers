```python
# prompt: Applique your_scaler.pkl sur df avec sklearn.preprocessing.Transformer

import pandas as pd
from sklearn.preprocessing import StandardScaler
import joblib

# Load the scaler
try:
    your_scaler = joblib.load('/content/sample_data/your_scaler.pkl') # Replace with the actual path to your scaler file
except FileNotFoundError:
    print("Error: 'your_scaler.pkl' not found. Please check the file path.")
    exit() # Exit the script if the scaler file is not found
except Exception as e:
    print(f"An error occurred while loading the scaler: {e}")
    exit()


# Replace 'Donées de prédictions.xlsx' with the actual file name if different
# Replace '/content/drive/MyDrive/...' with the actual path to your file in Google Drive
try:
    df = pd.read_excel('/content/sample_data/Données de prédictions.xlsx')

    # Assuming the 'Années' column contains the year information
    df['Trimestre'] = (df['Années'].dt.quarter) # Extract quarter from the date

    # Create seasonal dummy variables
    df['Saison_T1'] = (df['Trimestre'] == 1).astype(int)
    df['Saison_T2'] = (df['Trimestre'] == 2).astype(int)
    df['Saison_T3'] = (df['Trimestre'] == 3).astype(int)
    df['Saison_T4'] = (df['Trimestre'] == 4).astype(int)

    # Convert 'Covid-19' and 'Reprise du trafic' to boolean (0 or 1)
    # Replace 'Covid-19' and 'Reprise du trafic' with the actual column names if they are different
    df['Covid-19'] = df['Covid-19'].astype(int)  # Convert to integer (0 or 1)
    df['Reprise du trafic'] = df['Reprise du trafic'].astype(int) # Convert to integer (0 or 1)

    # Select columns to scale (excluding non-numeric columns and the target variable)
    columns_to_scale = ['Passagers', 'PIB Réel', 'Coût mondial du pétrole', 'Exportations', 'Importations','PIB par habitant', 'Saison_T1', 'Saison_T2', 'Saison_T3', 'Saison_T4', 'Covid-19', 'Reprise du trafic']

    # Apply the scaler
    df_scaled = pd.DataFrame(your_scaler.transform(df[columns_to_scale]), columns=columns_to_scale, index=df.index)

    #Concatenate scaled features with original dataframe
    df = pd.concat([df.drop(columns_to_scale, axis=1), df_scaled], axis=1)

    print(df.head())

except FileNotFoundError:
    print("Error: File not found. Please check the file path.")
except KeyError as e:
    print(f"Error: Column '{e}' not found in the DataFrame. Please check the column name.")
except AttributeError:
    print("Error: 'Années' column does not appear to contain datetime objects. Please ensure correct data type.")
except ValueError as e:
    print(f"Value Error: {e}") # Handle potential ValueErrors during scaling
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# prompt: Applique l'inverse de your_scale.pkl à df avec sklearn.preprocessing.Transformer

import pandas as pd
from sklearn.preprocessing import StandardScaler
import joblib

# Load the scaler
try:
    your_scaler = joblib.load('/content/sample_data/your_scaler.pkl')
except FileNotFoundError:
    print("Error: 'your_scaler.pkl' not found. Please check the file path.")
    exit()
except Exception as e:
    print(f"An error occurred while loading the scaler: {e}")
    exit()

# Assuming df is already loaded and preprocessed as in your original code
# ... (your code to load and preprocess df)

# Select columns to inverse transform
columns_to_inverse_transform = ['Passagers', 'PIB Réel', 'Coût mondial du pétrole', 'Exportations', 'Importations','PIB par habitant', 'Saison_T1', 'Saison_T2', 'Saison_T3', 'Saison_T4', 'Covid-19', 'Reprise du trafic']

# Apply the inverse transform
try:
    df_inverse_transformed = pd.DataFrame(your_scaler.inverse_transform(df[columns_to_inverse_transform]), columns=columns_to_inverse_transform, index=df.index)

    # Concatenate inverse transformed features with original dataframe
    df = pd.concat([df.drop(columns_to_inverse_transform, axis=1), df_inverse_transformed], axis=1)
    print(df.head())
except ValueError as e:
    print(f"Value Error during inverse transform: {e}")
except Exception as e:
    print(f"An error occurred during inverse transform: {e}")
```
