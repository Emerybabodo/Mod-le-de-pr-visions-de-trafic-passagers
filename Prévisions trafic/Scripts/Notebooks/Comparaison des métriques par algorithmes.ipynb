```python
import matplotlib.pyplot as plt
import numpy as np

# Données des modèles
models = [
    "SVR",
    "Ridge Regression",
    "Lasso Regression",
    "Elastic Net Regression",
    "Gradient Boosting Machine",
    "Bayesian Ridge Regression",
    "Modèle combiné"
]

metrics = {
    "MSE": [0.19, 0.14, 0.12, 0.24, 0.09, 0.31, 0.12],
    "RMSE": [0.43, 0.37, 0.35, 0.49, 0.30, 0.56, 0.34],
    "MAE": [0.38, 0.31, 0.29, 0.42, 0.10, 0.46, 0.27],
    "MAPE": [148.81, 67.01, 72.70, 129.93, 0.13, 1.46, 1.42],
    "R^2": [0.85, 0.89, 0.90, 0.82, 0.91, 0.69, 0.77]
}

# Création d'un graphique pour chaque métrique
fig, axes = plt.subplots(3, 2, figsize=(15, 12))
fig.tight_layout(pad=5)

metric_names = list(metrics.keys())
for i, ax in enumerate(axes.flat):
    if i < len(metric_names):
        metric = metric_names[i]
        values = metrics[metric]
        ax.bar(models, values, color='skyblue', edgecolor='black')
        ax.set_title(f"{metric} par modèle")
        ax.set_ylabel(metric)
        ax.set_xticklabels(models, rotation=45, ha='right')
    else:
        ax.axis('off')  # Désactive les sous-graphiques vides

plt.suptitle("Comparaison des métriques des différents modèles", fontsize=16, y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Données des modèles
models = [
    "SVR",
    "Ridge Regression",
    "Lasso Regression",
    "Elastic Net Regression",
    "Gradient Boosting Machine",
    "Bayesian Ridge Regression",
    "Combined Model"
]

metrics = {
    "MSE": [0.19, 0.14, 0.12, 0.24, 0.09, 0.31, 0.12],
    "RMSE": [0.43, 0.37, 0.35, 0.49, 0.30, 0.56, 0.34],
    "MAE": [0.38, 0.31, 0.29, 0.42, 0.10, 0.46, 0.27],
    "MAPE": [148.81, 67.01, 72.70, 129.93, 0.13, 1.46, 1.42],
    "R-squared": [0.85, 0.89, 0.90, 0.82, 0.91, 0.69, 0.77]
}

# Préparer le graphique
fig, axs = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle("Comparaison des métriques des différents modèles", fontsize=16)

# Couleurs pour les barres
colors = plt.cm.tab10(np.linspace(0, 1, len(models)))

# Graphique pour chaque métrique
for ax, (metric, values) in zip(axs.flatten(), metrics.items()):
    ax.bar(models, values, color=colors)
    ax.set_title(metric)
    ax.set_xticklabels(models, rotation=45, ha='right')
    ax.set_ylabel(metric)
    ax.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustement des espacements
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
