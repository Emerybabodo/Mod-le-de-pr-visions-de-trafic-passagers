```python
# prompt: mets toutes les valeurs des données sans chiffres après la virgule

import pandas as pd


def load_and_display_excel_data(filepath):
  """Loads data from an Excel file, converts numeric values to integers, and displays it.

  Args:
    filepath: The path to the Excel file.

  Returns:
    A pandas DataFrame containing the data from the Excel file, or None if
    an error occurs.
  """
  try:
    df = pd.read_excel(filepath)

    # Iterate through columns and convert numeric values to integers
    for col in df.columns:
      if pd.api.types.is_numeric_dtype(df[col]):
        df[col] = df[col].astype(int)

    print(f"Data from {filepath}:\n{df}\n")
    return df
  except FileNotFoundError:
    print(f"Error: File not found at {filepath}")
    return None
  except Exception as e:
    print(f"An error occurred while loading {filepath}: {e}")
    return None

# List of Excel files to load
excel_files = [
    "/content/sample_data/Bayesian ridge Plot.xlsx",
    "/content/sample_data/Elastic net Plot.xlsx",
    "/content/sample_data/Gradient Boosting Plot.xlsx",
    "/content/sample_data/Modèle combiné Plot.xlsx",
    "/content/sample_data/Ridge Plot.xlsx",
    "/content/sample_data/SVR Plot.xlsx",
    "/content/sample_data/Lasso Plot.xlsx"
]


# Load and display data from each file
dataframes = []  # Store DataFrames for later use if needed
for file in excel_files:
    df = load_and_display_excel_data(file)
    if df is not None:
        dataframes.append(df)
```
