```python
# prompt: Applique un test t au données de la varibale Passagers en comparant les données de la période du 01/03/2016 à 01/03/2020 aux données de la période 01/12/2020 à 01/09/2024. Affiche les résultats à savoir la moyenne, l"écart type, la variance, la médiane.

import pandas as pd
from scipy import stats

try:
    df = pd.read_excel('/content/sample_data/Global Trimestre.xlsx')

    # Convert 'Années' column to datetime objects
    df['Années'] = pd.to_datetime(df['Années'])

    # Filter data for the specified periods
    start_date1 = pd.to_datetime('2016-03-01')
    end_date1 = pd.to_datetime('2020-03-01')
    period1_data = df[(df['Années'] >= start_date1) & (df['Années'] <= end_date1)]['Passagers']

    start_date2 = pd.to_datetime('2020-12-01')
    end_date2 = pd.to_datetime('2024-09-01')
    period2_data = df[(df['Années'] >= start_date2) & (df['Années'] <= end_date2)]['Passagers']

    # Perform t-test
    t_statistic, p_value = stats.ttest_ind(period1_data, period2_data)

    # Calculate descriptive statistics
    def describe_data(data, period_name):
        mean = data.mean()
        std = data.std()
        variance = data.var()
        median = data.median()
        print(f"Descriptive Statistics for {period_name}:")
        print(f"  Mean: {mean}")
        print(f"  Standard Deviation: {std}")
        print(f"  Variance: {variance}")
        print(f"  Median: {median}")
        print("-" * 20)
        return mean, std, variance, median

    mean1, std1, variance1, median1 = describe_data(period1_data, "Period 1 (2016-03-01 to 2020-03-01)")
    mean2, std2, variance2, median2 = describe_data(period2_data, "Period 2 (2020-12-01 to 2024-09-01)")


    print(f"T-test results:")
    print(f"  T-statistic: {t_statistic}")
    print(f"  P-value: {p_value}")

except FileNotFoundError:
    print("Error: File 'Global Trimestre.xlsx' not found. Please check the file path.")
except KeyError as e:
    print(f"Error: Column '{e}' not found in the DataFrame.")
except Exception as e:
    print(f"An error occurred: {e}")
```

T-statistic : La statistique t de -0.58 indique que la différence entre les moyennes des deux groupes (avant et après Covid) est relativement faible par rapport à la variation au sein des groupes. Une valeur proche de 0 suggère que les moyennes sont très similaires.

P-value : La p-value de 0.5662 est bien supérieure à un seuil de signification classique de 0.05. Cela signifie que nous ne pouvons pas rejeter l'hypothèse nulle, qui stipule qu'il n'y a pas de différence significative entre les moyennes des deux groupes.

Conclusion :
Aucune différence significative : Avec une p-value de 0.5662, vous ne rejetez pas l'hypothèse nulle.
