```python
# prompt: Je veux prédire la variable passagers_diff à l'aide d'un modèle SARIMA-Robust qui prend en compte la tendance, la saisonnalité,  les anomalies Covid-19 et Reprise du trafic. divise les données de la valeur cible en données d'entrainement 80% et données test 20% Et tient compte de l'index des dates ci dessus avec freq='Q'

import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
import numpy as np

# Assuming 'df' is your DataFrame with 'Passagers_diff_stabilized' and a DatetimeIndex

# Split data into training and testing sets
train_size = int(len(df) * 0.8)
train_data = df['Passagers_diff_stabilized'][:train_size]
test_data = df['Passagers_diff_stabilized'][train_size:]

# Define the SARIMA model
# You need to find optimal (p, d, q) and (P, D, Q, s) orders
# Example: (p, d, q) = (1, 1, 1) and (P, D, Q, s) = (1, 1, 1, 4)
# s = 4 for quarterly data
model = SARIMAX(train_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 4))

# Fit the model
results = model.fit()


# Make predictions on the test set
predictions = results.get_prediction(start=len(train_data), end=len(df)-1)

# Extract predicted values
predicted_values = predictions.predicted_mean

# Evaluate the model
rmse = np.sqrt(mean_squared_error(test_data, predicted_values))
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Print predictions and actual values for comparison
print("Predictions:", predicted_values)
print("Actual Values:", test_data.values)
```

```python
# prompt: effectue la cross validation et affiche les résultats sur chaque folds

from sklearn.model_selection import TimeSeriesSplit
# ... (Your existing code)

# Assuming 'df' is your DataFrame and 'Passagers_diff_stabilized' is the stabilized time series data.

# Define the number of splits for time series cross-validation
n_splits = 5  # You can adjust this as needed
tscv = TimeSeriesSplit(n_splits=n_splits)

# Initialize lists to store results
rmse_scores = []
r2_scores = []
mse_scores = []
mae_scores = []

# Loop through the cross-validation splits
for train_index, test_index in tscv.split(df):
    train_data = df['Passagers_diff_stabilized'].iloc[train_index]
    test_data = df['Passagers_diff_stabilized'].iloc[test_index]

    # Fit the SARIMA model
    model = SARIMAX(train_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 4))
    results = model.fit(disp=False)

    # Make predictions
    predictions = results.get_forecast(steps=len(test_data))
    predicted_mean = predictions.predicted_mean

    # Calculate metrics
    rmse = np.sqrt(mean_squared_error(test_data, predicted_mean))
    r2 = r2_score(test_data, predicted_mean)
    mse = mean_squared_error(test_data, predicted_mean)
    mae = mean_absolute_error(test_data, predicted_mean)

    # Store the results
    rmse_scores.append(rmse)
    r2_scores.append(r2)
    mse_scores.append(mse)
    mae_scores.append(mae)

    # Print the results for the current fold
    print(f"Fold Results (Fold {len(rmse_scores)}):")
    print(f"RMSE: {rmse}")
    print(f"R-squared (R2): {r2}")
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"Mean Absolute Error (MAE): {mae}")
    print("-" * 20)

# Print the overall average results
print(f"\nAverage Results (across {n_splits} folds):")
print(f"Average RMSE: {np.mean(rmse_scores)}")
print(f"Average R-squared (R2): {np.mean(r2_scores)}")
print(f"Average Mean Squared Error (MSE): {np.mean(mse_scores)}")
print(f"Average Mean Absolute Error (MAE): {np.mean(mae_scores)}")
```
